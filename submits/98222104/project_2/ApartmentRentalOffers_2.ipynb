# Pre Processing

## Find the null data

```python
df.isna().sum()
```

```python
df.isna().sum()/len(df)
```


### Delete columns with more than 50% null data

```python
null_columns = df.columns[((df.isna().sum()/len(df)) > 0.50)]
null_columns
```

```python
df = df.drop(columns = null_columns)
df.columns
```


### Delete columns without useful information

```python
df = df.drop(columns=['facilities','description','livingSpaceRange','scoutId','street',
                      'streetPlain','houseNumber','date','geo_bln','geo_krs','geo_plz',
                      'picturecount','telekomTvOffer'])
```

```python
df.info()
```

### Fillna numeric data by mean

```python
df.fillna(df._get_numeric_data().mean(), inplace = True)
```

```python
df.isna().sum()
```

### Fillna categorical data

```python
for col in df.columns:
    if df[col].dtype == 'object' or df[col].dtype == 'bool':
      mode = df[col].value_counts().head(1).index[0]
      print('Null data in \"{}\" column fill with \n\"{}\"\n'.format(col , mode))
      df[col].fillna(mode, inplace = True)
```

```python
df.isna().sum()
```

## Find incorrect data

```python
df[df['totalRent'] == 0.0].shape[0]
```

```python
df = df.drop(df[df['totalRent'] == 0.0].index)
df.shape
```

```python
df = df.drop(df[df['livingSpace'] == 0.0].index)
df.shape
```

## Find the outlier data

```python
df.dtypes
```

```python
for cols in df.columns:
    if df[cols].dtype == 'int64' or df[cols].dtype == 'float64':
        upper_range = df[cols].mean() + 3 * df[cols].std()
        lower_range = df[cols].mean() - 3 * df[cols].std()
        
        indexs = df[(df[cols] > upper_range) | (df[cols] < lower_range)].index
        df = df.drop(indexs)
```

```python
df.shape
```

## Find categorical features with lots of states

```python
for cols in df.columns:
    if df[cols].dtype == 'object' or df[cols].dtype == 'bool':
        print('{} : {}'.format(cols,df[cols].nunique()))
```

```python
df = df.drop(columns=['firingTypes','regio2','regio3'])
df.shape
```

## Find duplicate rows

```python
df.duplicated().value_counts()
```

```python
df.drop_duplicates(inplace = True)
df.shape
```

```python
df.head()
```

```python
df1 = df[['serviceCharge', 'heatingType', 'telekomUploadSpeed', 'totalRent']]
df1.head()
```

```python
df1 = pd.get_dummies(df1)
df1.head()
```

```python
X = df1.drop(['totalRent'], axis=1)
y = df1['totalRent']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_train_scaled = pd.DataFrame(X_train_scaled, columns = X_train.columns)
X_test_scaled= scaler.transform(X_test)
X_test_scaled = pd.DataFrame(X_test_scaled, columns = X_test.columns)
```

```python
class LinearRegression():
  def __init__(self, columns_size, iteration, batch_size, learning_rate):
    self.weights = np.random.rand(columns_size)
    self.iteration = iteration
    self.batch_size = batch_size
    self.learning_rate = learning_rate

  def batching(self, x, y):
    batches = []
    size = x.shape[0]
    indexes = list(range(size))
    np.random.shuffle(indexes)
    for i in range(0, size, self.batch_size):
      j = i + self.batch_size
      j = j if j < size else size
      batch = indexes[i:j]
      batches.append((x.take(batch, axis = 0), y.take(batch)))
    return batches

  def loss(self, x, y, yp):
    return (np.dot(x.T, yp - y))/(len(x))
  
  def predict(self, x):
    return np.dot(x, self.weights)

  def mse(self, yp, y):
    diff = yp - y
    se = np.power(diff, 2, dtype='float64')
    return np.mean(se)

  def train(self, X_train, y_train):
    batches = self.batching(X_train, y_train)
    for i in range(self.iteration):
      for batch in batches:
        x, y = batch
        yp = self.predict(X_train)
        loss = self.loss(X_train, y_train, yp)
        self.weights = self.weights - self.learning_rate * loss

    return self.weights
```

```python
LR = LinearRegression(X_train_scaled.shape[1], 1000, 50000, 0.1)
LR.train(X_train_scaled, y_train)
y_pred = LR.predict(X_test_scaled)
result = pd.DataFrame({'test':y_test,'prediction':y_pred})
result
```

```python
LR.mse(y_pred, y_test)
```
